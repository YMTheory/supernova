```python

arrNO_1DnoC14 = np.loadtxt("../tmp_data/Garching82703_10kpc_NO_pESeESIBD_MOSensAsimovnoC14_JUNO.csv")
dTNO_pdfNO_1DnoC14 = arrNO_1DnoC14[:, 1]
nllNO_pdfNO_1DnoC14 = arrNO_1DnoC14[:, 2]
dTNO_pdfIO_1DnoC14 = arrNO_1DnoC14[:, 3]
nllNO_pdfIO_1DnoC14 = arrNO_1DnoC14[:, 4]

arrIO_1DnoC14 = np.loadtxt("../tmp_data/Garching82703_10kpc_IO_pESeESIBD_MOSensAsimovnoC14_JUNO.csv")
dTIO_pdfNO_1DnoC14 = arrIO_1DnoC14[:, 1]
nllIO_pdfNO_1DnoC14 = arrIO_1DnoC14[:, 2]
dTIO_pdfIO_1DnoC14 = arrIO_1DnoC14[:, 3]
nllIO_pdfIO_1DnoC14 = arrIO_1DnoC14[:, 4]

arrNO_1D = np.loadtxt("../tmp_data/Garching82703_10kpc_NO_pESeESIBD_MOSensAsimovC14low_JUNO.csv")
dTNO_pdfNO_1D = arrNO_1D[:, 1]
nllNO_pdfNO_1D = arrNO_1D[:, 2]
dTNO_pdfIO_1D = arrNO_1D[:, 3]
nllNO_pdfIO_1D = arrNO_1D[:, 4]

arrIO_1D = np.loadtxt("../tmp_data/Garching82703_10kpc_IO_pESeESIBD_MOSensAsimovC14low_JUNO.csv")
dTIO_pdfNO_1D = arrIO_1D[:, 1]
nllIO_pdfNO_1D = arrIO_1D[:, 2]
dTIO_pdfIO_1D = arrIO_1D[:, 3]
nllIO_pdfIO_1D = arrIO_1D[:, 4]

arrNO_pES2D = np.loadtxt("../tmp_data/Garching82703_10kpc_NO_pES2DeESIBD_MOSensAsimovC14low_JUNO.csv")
dTNO_pdfNO_pES2D = arrNO_pES2D[:, 1]
nllNO_pdfNO_pES2D = arrNO_pES2D[:, 2]
dTNO_pdfIO_pES2D = arrNO_pES2D[:, 3]
nllNO_pdfIO_pES2D = arrNO_pES2D[:, 4]

arrIO_pES2D = np.loadtxt("../tmp_data/Garching82703_10kpc_IO_pES2DeESIBD_MOSensAsimovC14low_JUNO.csv")
dTIO_pdfNO_pES2D = arrIO_pES2D[:, 1]
nllIO_pdfNO_pES2D = arrIO_pES2D[:, 2]
dTIO_pdfIO_pES2D = arrIO_pES2D[:, 3]
nllIO_pdfIO_pES2D = arrIO_pES2D[:, 4]

arrNO_ES2D = np.loadtxt("../tmp_data/Garching82703_10kpc_NO_pES2DeES2DIBD_MOSensAsimovC14low_JUNO.csv")
dTNO_pdfNO_ES2D = arrNO_ES2D[:, 1]
nllNO_pdfNO_ES2D = arrNO_ES2D[:, 2]
dTNO_pdfIO_ES2D = arrNO_ES2D[:, 3]
nllNO_pdfIO_ES2D = arrNO_ES2D[:, 4]

arrIO_ES2D = np.loadtxt("../tmp_data/Garching82703_10kpc_IO_pES2DeES2DIBD_MOSensAsimovC14low_JUNO.csv")
dTIO_pdfNO_ES2D = arrIO_ES2D[:, 1]
nllIO_pdfNO_ES2D = arrIO_ES2D[:, 2]
dTIO_pdfIO_ES2D = arrIO_ES2D[:, 3]
nllIO_pdfIO_ES2D = arrIO_ES2D[:, 4]
```

```python


Ethr = np.arange(0, 160, 10)
mode = "nll"
fig, ax = plt.subplots(1, 2, figsize=(10, 4))
if mode == "dT":
    ax[0].plot(Ethr, dTNO_pdfIO_1DnoC14, "-.", lw=2.5, label="1D, no C14")
    ax[0].plot(Ethr, dTNO_pdfIO_1D,      "-",  lw=2.5, label="1D")
    ax[0].plot(Ethr, dTNO_pdfIO_pES2D,   "--", lw=2.5, label="pES 2D")
    ax[0].plot(Ethr, dTNO_pdfIO_ES2D,    ":",  lw=2.5, label="pES+eES 2D")
    ax[0].set_ylabel(r"$\Delta t$ [ms]", fontsize=15)
    ax[0].legend(prop={"size":12})
    ax[0].grid(True)
    ax[0].set_title("NO", fontsize=16)
    ax[0].tick_params(axis="both", labelsize=15)
    ax[1].plot(Ethr, dTIO_pdfNO_1DnoC14, "-.", lw=2.5, label="1D, no C14")
    ax[1].plot(Ethr, dTIO_pdfNO_1D,      "-",  lw=2.5, label="1D")
    ax[1].plot(Ethr, dTIO_pdfNO_pES2D,   "--", lw=2.5, label="pES 2D")
    ax[1].plot(Ethr, dTIO_pdfNO_ES2D,    ":",  lw=2.5, label="pES+eES 2D")
    ax[1].set_ylabel(r"$\Delta t$ [ms]", fontsize=15)
    ax[1].legend(prop={"size":12})
    ax[1].grid(True)
    ax[1].set_title("IO", fontsize=16)
    ax[1].tick_params(axis="both", labelsize=15)
elif mode == "nll":
    ax[0].plot(Ethr, 2*nllNO_pdfIO_1DnoC14 - 2*nllNO_pdfNO_1DnoC14,       "-",  lw=2.5, label="1D, no "+r"$^{14}$C")
    ax[0].plot(Ethr, 2*nllNO_pdfIO_1D - 2*nllNO_pdfNO_1D,       "-",  lw=2.5, label="1D, "+r"$1\times10^{-18}$g/g$^{14}$C")
    ax[0].plot(Ethr, 2*nllNO_pdfIO_pES2D - 2*nllNO_pdfNO_pES2D, "--", lw=2.5, label="pES 2D, "+r"$1\times10^{-18}$g/g$^{14}$C")
    ax[0].plot(Ethr, 2*nllNO_pdfIO_ES2D - 2*nllNO_pdfNO_ES2D,   ":",  lw=2.5, label="pES+eES 2D, "+r"$1\times10^{-18}$g/g$^{14}$C")
    ax[0].set_ylabel(r"$\Delta \chi^2$", fontsize=15)
    ax[0].set_title("NO", fontsize=16)
    ax[0].tick_params(axis="both", labelsize=15)
    ax[0].grid(True)
    ax[0].legend(prop={"size":12})
    ax[0].set_ylim(6.5, 11)
    ax[1].plot(Ethr, 2*nllIO_pdfNO_1DnoC14 - 2*nllIO_pdfIO_1DnoC14,       "-",  lw=2.5, label="1D, no C14")
    ax[1].plot(Ethr, 2*nllIO_pdfNO_1D - 2*nllIO_pdfIO_1D,       "-",  lw=2.5, label="1D")
    ax[1].plot(Ethr, 2*nllIO_pdfNO_pES2D - 2*nllIO_pdfIO_pES2D, "--", lw=2.5, label="pES 2D")
    ax[1].plot(Ethr, 2*nllIO_pdfNO_ES2D - 2*nllIO_pdfIO_ES2D,   ":",  lw=2.5, label="pES+eES 2D")
    ax[1].set_ylabel(r"$\Delta \chi^2$", fontsize=15)
    ax[1].grid(True)
    ax[1].set_title("IO", fontsize=16)
    ax[1].tick_params(axis="both", labelsize=15)
    ax[1].set_ylim(6.5, 11)
    #ax[1].legend(prop={"size":12})
plt.tight_layout()
plt.savefig(f"../tmp_plots/Garching82703_10kpc_lowC14_{mode}_differentFitter.pdf")
```
