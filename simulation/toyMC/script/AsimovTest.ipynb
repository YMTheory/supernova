```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def loadAsimovResults(filename, idx, num):
    arr = np.loadtxt(filename)
    res = [None for i in range(num)]
    for i in range(num):
        res[i] = arr[:, idx+i]
    return res
```

```python
path = "/junofs/users/miaoyu/supernova/simulation/toyMC/results/"
prefix = "Garching82703_variedDist_"
suffix = "MOSensAsimov_CombSep.csv"

exp = "SuperK_"
cha = "eESIBD_"
dTNO_SK_totcomb, dchi2NO_SK_totcomb, dTNO_SK_part, dchi2NO_SK_part, dTNO_SK_sep, dchi2NO_SK_sep = loadAsimovResults(path+prefix+exp+"NO_"+cha+suffix, 1, 6)
dTIO_SK_totcomb, dchi2IO_SK_totcomb, dTIO_SK_part, dchi2IO_SK_part, dTIO_SK_sep, dchi2IO_SK_sep = loadAsimovResults(path+prefix+exp+"IO_"+cha+suffix, 1, 6)

exp = "HyperK_"
cha = "eESIBDshort_"
dTNO_HK_totcomb, dchi2NO_HK_totcomb, dTNO_HK_part, dchi2NO_HK_part, dTNO_HK_sep, dchi2NO_HK_sep = loadAsimovResults(path+prefix+exp+"NO_"+cha+suffix, 1, 6)
dTIO_HK_totcomb, dchi2IO_HK_totcomb, dTIO_HK_part, dchi2IO_HK_part, dTIO_HK_sep, dchi2IO_HK_sep = loadAsimovResults(path+prefix+exp+"IO_"+cha+suffix, 1, 6)

exp = "JUNO_"
cha = "eESIBD_"
dTNO_JUNO_totcomb, dchi2NO_JUNO_totcomb, dTNO_JUNO_part, dchi2NO_JUNO_part, dTNO_JUNO_sep, dchi2NO_JUNO_sep = loadAsimovResults(path+prefix+exp+"NO_"+cha+suffix, 1, 6)
dTIO_JUNO_totcomb, dchi2IO_JUNO_totcomb, dTIO_JUNO_part, dchi2IO_JUNO_part, dTIO_JUNO_sep, dchi2IO_JUNO_sep = loadAsimovResults(path+prefix+exp+"IO_"+cha+suffix, 1, 6)
```

```python
dist = np.arange(5, 11, 1)
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

ax[0].plot(dist, dchi2NO_SK_totcomb, "o-", color="blue", lw=2.5, label="SuperK: combined")
ax[0].plot(dist, dchi2NO_SK_part, "o:", color="blue", lw=2.5, label="SuperK: IBD tag eff.= 50%")
ax[0].plot(dist, dchi2NO_SK_sep, "o-.", color="blue", lw=2.5, label="SuperK: separately")
ax[0].plot(dist, dchi2NO_HK_totcomb, "o-", color="darkorange", lw=2.5, label="HyperK: IBD tag eff. = 0%")
#ax[0].plot(dist, dchi2NO_HK_part, "o:", color="darkorange", lw=2.5, label="HyperK: IBD tag eff.= 50%")
#ax[0].plot(dist, dchi2NO_HK_sep, "o-.", color="darkorange", lw=2.5, label="HyperK: separately")
ax[0].set_xlabel("Distance [kpc]", fontsize=18)
ax[0].set_ylabel(r"$\Delta \chi^2$", fontsize=18)
ax[0].grid(True)
#ax[0].legend(prop={"size":16})
ax[0].tick_params(axis="both", labelsize=15)
ax[0].set_title("NO", fontsize=18)
#ax[0].set_ylim(0, 20)

ax[1].plot(dist, dchi2IO_SK_totcomb, "o-", color="blue", lw=2.5, label="SuperK: IBD tag eff.= 0%")
ax[1].plot(dist, dchi2IO_SK_part, "o:", color="blue", lw=2.5, label="SuperK: IBD tag eff.= 50%")
ax[1].plot(dist, dchi2IO_SK_sep, "o-.", color="blue", lw=2.5, label="SuperK: IBD tag eff. = 100%")
ax[1].plot(dist, dchi2IO_HK_totcomb, "o-", color="darkorange", lw=2.5, label="HyperK: IBD tag eff. = 0%")
#ax[1].plot(dist, dchi2IO_HK_part, "o:", color="darkorange", lw=2.5, label="HyperK: IBD tag eff.= 50%")
#ax[1].plot(dist, dchi2IO_HK_sep, "o-.", color="darkorange", lw=2.5, label="HyperK: separately")
ax[1].set_xlabel("Distance [kpc]", fontsize=18)
ax[1].set_ylabel(r"$\Delta \chi^2$", fontsize=18)
ax[1].grid(True)
ax[1].legend(prop={"size":16})
ax[1].tick_params(axis="both", labelsize=15)
ax[1].set_title("IO", fontsize=18)
#ax[1].set_ylim(0, 20)

plt.tight_layout()
plt.savefig("../plots/Garching82703_variedDist_SuperK3casesHyperK_AsimovSens.pdf")
```
