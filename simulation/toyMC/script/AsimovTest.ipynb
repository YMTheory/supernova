```python
import re

NOimg, IOimg = np.zeros((6, 6)), np.zeros((6, 6))
row, col = 5, 0
for dist in [5, 6, 7, 8, 9, 10]:
    col = 0
    for mass in [0.5, 1.0, 1.5, 2.0, 2.5, 3.0]:
        NOfilename = f"../tmp_data/NOlowC14Ethr0.10MeV_target{mass}_dist{dist}kpc_pES2D.txt"
        with open(NOfilename) as fNO:
            for lines in fNO.readlines():
                line = lines.strip("\n")
                if re.findall("Output", line):
                    data = line.split(" ")
                    NOimg[row, col] = 2 * (float(data[5]) - float(data[3]))
                    
        IOfilename = f"../tmp_data/IOlowC14Ethr0.10MeV_target{mass}_dist{dist}kpc_pES2D.txt"
        with open(IOfilename) as fIO:
            for lines in fIO.readlines():
                line = lines.strip("\n")
                if re.findall("Output", line):
                    data = line.split(" ")
                    IOimg[row, col] = -2 * (float(data[5]) - float(data[3]))
    
        col = col + 1
    row = row - 1
    
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 4))

im0 = ax[0].imshow(NOimg, extent=[5, 65, 4.5, 10.5], aspect="auto")
ax[0].set_title("NO", fontsize=18)
ax[0].set_xlabel("Target mass [kton]", fontsize=18)
ax[0].set_ylabel("Distance [kpc]", fontsize=18)
ax[0].tick_params(axis="both", labelsize=17)
x = np.arange(10, 70, 10)
y = [10, 9, 8, 7, 6, 5]
X, Y = np.meshgrid(x, y)
cs0 = ax[0].contour(X, Y, NOimg, levels=[9, 16, 25], cmap="cool")
ax[0].clabel(cs0, fontsize=18)
cb0 = plt.colorbar(im0, ax=ax[0])
cb0.set_label(r"$\Delta\chi^2$", fontsize=18)
cb0.ax.tick_params(labelsize=16)

im1 = ax[1].imshow(IOimg, extent=[5, 65, 4.5, 10.5], aspect="auto")
ax[1].set_title("IO", fontsize=18)
ax[1].set_xlabel("Target mass [kton]", fontsize=18)
ax[1].set_ylabel("Distance [kpc]", fontsize=18)
ax[1].tick_params(axis="both", labelsize=17)
x = np.arange(10, 70, 10)
y = [10, 9, 8, 7, 6, 5]
X, Y = np.meshgrid(x, y)
cs0 = ax[1].contour(X, Y, IOimg, levels=[9, 16, 25], cmap="cool")
ax[1].clabel(cs0, fontsize=18)
cb1 = plt.colorbar(im1, ax=ax[1])
cb1.set_label(r"$\Delta\chi^2$", fontsize=18)
cb1.ax.tick_params(labelsize=16)

plt.tight_layout()

plt.savefig("../tmp_plots/Garching82703_10kpc_Ethr0.1MeV_lowC14_AsimovSensDistTargetMass_2Dfit.pdf")
```
