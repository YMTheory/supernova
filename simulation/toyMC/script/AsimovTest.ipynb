```python
def loadAsimovResults(filename, idx, num):
    arr = np.loadtxt(filename)
    res = [None for i in range(num)]
    for i in range(num):
        res[i] = arr[:, idx+i]
    return res
```

```python
import re
def readSingleEvtOutput(filename):
    with open(filename, "r") as f:
        for lines in f.readlines():
            line = lines.strip("\n")
            if re.findall("Output", line):
                data = line.split(" ")
                dT_NOfit = float(data[2])
                nll_NOfit = float(data[3])
                dT_IOfit = float(data[4])
                nll_IOfit = float(data[5])
    return dT_NOfit, nll_NOfit, dT_IOfit, nll_IOfit

```

```python
dist_arr = np.arange(5, 10.1, 0.1)
mass_arr = np.arange(0.1, 3.6, 0.1)
NOarr, IOarr = np.zeros((len(mass_arr), len(dist_arr))), np.zeros((len(mass_arr), len(dist_arr)))
i, j = len(mass_arr)-1, 0
for target in mass_arr:
    j = 0
    for dist in dist_arr:
        filename = f"../jobs/NOlowC14Ethr0.10MeV_target{target:.1f}_dist{dist:.1f}kpc_pES2D.txt"
        if filename == "../jobs/NOlowC14Ethr0.10MeV_target0.7_dist8.0kpc_pES2D.txt":
            filename = "../jobs/NOlowC14Ethr0.10MeV_target0.7_dist7.9kpc_pES2D.txt"
        _, nllNO, _, nllIO = readSingleEvtOutput(filename)
        NOarr[i, j] = 2 * nllIO - 2 * nllNO
        filename = f"../jobs/IOlowC14Ethr0.10MeV_target{target:.1f}_dist{dist:.1f}kpc_pES2D.txt"
        if filename == "../jobs/IOlowC14Ethr0.10MeV_target0.3_dist7.6kpc_pES2D.txt":
            filename = "../jobs/IOlowC14Ethr0.10MeV_target0.3_dist7.5kpc_pES2D.txt"
        elif filename == "../jobs/IOlowC14Ethr0.10MeV_target1.3_dist6.4kpc_pES2D.txt":
            filename = "../jobs/IOlowC14Ethr0.10MeV_target1.3_dist6.3kpc_pES2D.txt"
        _, nllNO, _, nllIO = readSingleEvtOutput(filename)
        IOarr[i, j] = 2 * nllNO - 2 * nllIO

        j += 1
    i = i - 1
```

```python

X, Y = np.meshgrid(dist_arr, mass_arr[::-1]*20)

fig, ax = plt.subplots(1, 2, figsize=(12, 5))
im0 = ax[0].imshow(NOarr, extent=[5, 10, 2, 70], aspect="auto", cmap="viridis")
ax[0].set_xlabel("Distacen [kpc]", fontsize=18)
ax[0].set_ylabel("Target mass [kton]", fontsize=18)
ax[0].tick_params(axis="both", labelsize=16)
cb0 = plt.colorbar(im0, ax=ax[0])
cb0.set_label(r"$\Delta\chi^2$", fontsize=18)
cb0.ax.tick_params(labelsize=16)
ct0 = ax[0].contour(X, Y, NOarr, levels=[9, 16, 25], cmap="cool")
plt.clabel(ct0, fontsize=16)
ax[0].set_title("NO", fontsize=18)

im1 = ax[1].imshow(NOarr, extent=[5, 10, 2, 70], aspect="auto", cmap="viridis")
ax[1].set_xlabel("Distacen [kpc]", fontsize=18)
ax[1].set_ylabel("Target mass [kton]", fontsize=18)
ax[1].tick_params(axis="both", labelsize=16)
cb1 = plt.colorbar(im1, ax=ax[1])
cb1.set_label(r"$\Delta\chi^2$", fontsize=18)
cb1.ax.tick_params(labelsize=16)
ct1 = ax[1].contour(X, Y, IOarr, levels=[9, 16, 25], cmap="cool")
plt.clabel(ct1, fontsize=16)
ax[1].set_title("IO", fontsize=18)

plt.tight_layout()
plt.savefig("../plots/Garching82703_variedDistvariedMass_JUNO_AsimovSensC14low.pdf")
```
