```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
plt.style.use("science")
```

```python
def parse_output(filename):
    df = pd.read_csv(filename)
    return df
```

```python
MO = "NO"
Ethr = 0.10

df = parse_output(f"./results/Garching82503_10kpc_{MO}_pESeESIBD_{Ethr:.2f}MeV.csv")
fig, ax = plt.subplots(1, 2, figsize=(8, 4))
ax[0].hist(df["TbestNO"], bins=50, histtype="step", label="NO");
ax[0].hist(df["TbestIO"], bins=50, histtype="step", label="IO");
ax[1].hist(df["sens"], bins=50, color="blue", edgecolor="black");
ax[0].legend(prop={"size": 13})
ax[0].set_xlabel(r"$\Delta T$ [ms]", fontsize=14)
ax[0].tick_params(axis="both", labelsize=14)

ax[1].vlines(np.median(df["sens"]), 0, 35, color="black", lw=2.5, linestyles="--");
med = np.median(df["sens"])
ax[1].text(0, 30, f"{med:.2f}", fontsize=13)
ax[1].tick_params(axis="both", labelsize=14)
ax[1].set_xlabel(r"$\Delta \chi^2$ [ms]", fontsize=14)

ax[0].set_title(f"Garching82503_10kpc_{MO}_pESeESIBD_{Ethr:.2f}MeV")
plt.tight_layout()
plt.savefig(f"./results/Garching82503_10kpc_{MO}_pESeESIBD_{Ethr:.2f}MeV_fineFit.pdf")
```
