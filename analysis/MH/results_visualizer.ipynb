```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
plt.style.use("science")
```

```python
def parse_output(filename):
    df = pd.read_csv(filename)
    return df
```

```python
MO = "IO"
Ethr = 0.15
model = "Garching82503"
fitTmax = 50

df = parse_output(f"./results/{model}_10kpc_{MO}_pESeESIBD_{Ethr:.2f}MeV_fitTmax{fitTmax}ms.csv")
fig, ax = plt.subplots(1, 2, figsize=(8, 4))
ax[0].hist(df["TbestNO"], bins=50, histtype="step", label="NO");
ax[0].hist(df["TbestIO"], bins=50, histtype="step", label="IO");
ax[1].hist(df["sens"], bins=50, color="blue", edgecolor="black");
ax[0].legend(prop={"size": 13})
ax[0].set_xlabel(r"$\Delta T$ [ms]", fontsize=14)
ax[0].tick_params(axis="both", labelsize=14)

ax[1].vlines(np.median(df["sens"]), 0, 35, color="black", lw=2.5, linestyles="--");
med = np.median(df["sens"])
ax[1].text(0, 30, f"{med:.2f}", fontsize=13)
ax[1].tick_params(axis="both", labelsize=14)
ax[1].set_xlabel(r"$\Delta \chi^2$ [ms]", fontsize=14)

ax[0].set_title(f"{model}_10kpc_{MO}_pESeESIBD_{Ethr:.2f}MeV")
plt.tight_layout()
plt.savefig(f"./results/{model}_10kpc_{MO}_pESeESIBD_{Ethr:.2f}MeV_fitTmax{fitTmax}ms_fineFit.pdf")
```

```python
MO = "NO"
Ethr = 0.15
model = "Garching82503"
fitTmax = 47

sens_NO1, sens_IO1 = [], []
for fitTmax in range(45, 56):
    df = parse_output(f"./results/{model}_10kpc_NO_pESeESIBD_{Ethr:.2f}MeV_fitTmax{fitTmax}ms.csv")
    sens = df["sens"]
    median_sens_NO = np.median(sens)
    sens_NO1.append(median_sens_NO)

    df = parse_output(f"./results/{model}_10kpc_IO_pESeESIBD_{Ethr:.2f}MeV_fitTmax{fitTmax}ms.csv")
    sens = df["sens"]
    median_sens_IO = np.median(sens)
    sens_IO1.append(median_sens_IO)


Ethr = 0.10
sens_NO2, sens_IO2 = [], []
for fitTmax in range(45, 56):
    df = parse_output(f"./results/{model}_10kpc_NO_pESeESIBD_{Ethr:.2f}MeV_fitTmax{fitTmax}ms.csv")
    sens = df["sens"]
    median_sens_NO = np.median(sens)
    sens_NO2.append(median_sens_NO)

    df = parse_output(f"./results/{model}_10kpc_IO_pESeESIBD_{Ethr:.2f}MeV_fitTmax{fitTmax}ms.csv")
    sens = df["sens"]
    median_sens_IO = np.median(sens)
    sens_IO2.append(median_sens_IO)
```

```python

def load_sens(MO, Ethr, eos, mass, trp):


    mod_num = (str(eos) + str(int(mass*10)) + str(trp))
    model = "Garching" + mod_num
    
    if mod_num == "82503":
        filename = f"./results/{model}_10kpc_{MO}_pESeESIBD_{Ethr:.2f}MeV_fitTmax50ms.csv"
    else:
        filename = f"./results/{model}_10kpc_{MO}_pESeESIBD_{Ethr:.2f}MeV_fitTmax20ms.csv"

    df = parse_output(filename)
    medium_sens = np.median(df["sens"])

    return int(mod_num), medium_sens
```

```python


ms = ["o", "*", "^", "v"]
ss = [8 ,10, 8, 8]

MO = "IO"
Ethr = 0.15

fig, ax = plt.subplots(figsize=(8, 6))

for  im, mass in enumerate([11.2, 25, 27, 40]):
    for trp in [0 ,1, 2, 3]:
        model, sens = load_sens(MO, Ethr, 8, mass, trp)
        ax.plot(im, sens, marker=ms[trp], ms=ss[trp], alpha=0.5, color="blue")
        model, sens = load_sens(MO, Ethr, 9, mass, trp)
        ax.plot(im, sens, marker=ms[trp], ms=ss[trp], color="coral", fillstyle="none")
        
ax.set_xlabel(r"Pregenitor mass [$M_\odot$]", fontsize=14)    
ax.set_ylabel(r"$\Delta \chi^2$", fontsize=14)
ax.set_xticks([0, 1, 2, 3], [11.2, 25, 27, 40])
ax.tick_params(axis="both", labelsize=14)
ax.set_xlim(-0.5, 3.5)
ax.set_ylim(5, 14)
#ax.set_ylim(2, 8)

ax.plot(-5, 5, ms[0], color="blue", label="Shen EoS")
ax.plot(-5, 5, ms[0], color="black", label="w/o Conv, w/o NP")
ax.plot(-5, 5, ms[1], color="black", label="w/ Conv, w/o NP")
ax.plot(-5, 5, ms[0], color="coral", fillstyle="none", label="LS220 EoS")
ax.plot(-5, 5, ms[2], color="black", label="w/o Conv, w/ NP")
ax.plot(-5, 5, ms[3], color="black", label="w/ Conv, w/ NP")
ax.legend(loc="lower left", ncol=2, prop={"size":15})
if MO == "NO":
    ax.hlines(4, -0.5, 3.5, linestyle="--", lw=1.5, color="gray")
    ax.text(1.8, 3.8, r"$2\sigma$", fontsize=15)
if MO == "IO":
    ax.hlines(9, -0.5, 3.5, linestyle="--", lw=1.5, color="gray")
    ax.text(1.3, 8.5, r"$3\sigma$", fontsize=15)
ax.set_title(f"{MO}_10kpc_Ethr{Ethr}MeV", fontsize=16)

plt.tight_layout()
plt.savefig(f"{MO}_10kpc_Ethr{Ethr}MeV_Garching.pdf")
```
